Q1  What is the difference between a function and a method in Python?

--Function: A standalone block of code that performs a specific task.
Method: A function that is associated with an object (class).

Q2 Explain the concept of function arguments and parameters in Python.

--Parameters: These are the variables defined within the parentheses of a function's definition. They act as placeholders for the values that will be passed to the function.
Arguments:  These are the actual values that are passed to the function when it is called. They are used to fill the placeholders (parameters) defined in the function's definition.

Q3) What are the different ways to define and call a function in Python?
--Defining: Use def function_name(parameters): followed by the function's code.
Calling: Use function_name(arguments) to execute the function, passing values for the parameters.


Q4)What is the purpose of the `return` statement in a Python function?
--To exit the function: When the return statement is encountered, the function immediately stops executing and returns control to the caller.

To provide a value back to the caller: The return statement can optionally be followed by an expression. The value of this expression is then returned to the caller of the function.

Q5)What are iterators in Python and how do they differ from iterables?
--Iterables: Objects you can loop over (e.g., lists, tuples). Provide a means to access their elements.
Iterators: Objects that implement the iterator protocol (__iter__ and __next__). Used to access elements of an iterable one by one.
Key Difference: Iterables can be looped over directly, while iterators are created from iterables and provide sequential access to their elements.

Q6)Explain the concept of generators in Python and how they are defined.
--Generators: Functions that produce a sequence of values using the yield keyword instead of return.
Benefits: Memory efficient, lazy evaluation, and concise for generating sequences.

Q7) What are the advantages of using generators over regular functions?
--Memory Efficient: Generate values on-demand, reducing memory usage.
Lazy Evaluation: Compute values as needed, improving performance, especially for large datasets.

Q8)What is a lambda function in Python and when is it typically used?
--Lambda: An anonymous function defined using lambda arguments: expression.
Uses: For short, single-line functions, callbacks in higher-order functions, and custom sorting logic.

Q9)Explain the purpose and usage of the `map()` function in Python.
--Purpose: Applies a given function to each item in an iterable (e.g., list, tuple).
Usage: map(function, iterable)


Q10)What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

--map(): Applies a function to each element of an iterable, returning a new iterable with the results.
filter(): Creates a new iterable with elements from the original that satisfy a given condition.
reduce(): Applies a function cumulatively to the elements of an iterable, reducing them to a single value.
